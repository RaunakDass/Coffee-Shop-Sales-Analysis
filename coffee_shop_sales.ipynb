FINDING TOTAL SALES IN JUNE

```python
june_transactions = df[df['transaction_date'].dt.month == 6]
june_transactions['total_sales'] = june_transactions['unit_price'] * june_transactions['transaction_qty']
total_sales_june = round(june_transactions['total_sales'].sum(), 2)
print(f"Total sales in June: {total_sales_june}")
```

FINDING TOTAL SALES IN JAN

```python
jan_transactions = df[df['transaction_date'].dt.month == 1]
jan_transactions['total_sales'] = jan_transactions['unit_price'] * jan_transactions['transaction_qty']
total_sales_jan = round(jan_transactions['total_sales'].sum(), 2)
print(f"Total sales in Jan: {total_sales_jan}")
```

FINDING MONTH ON MONTH SALES INCREASE/DECREASE

```python
df['total_sales'] = df['unit_price'] * df['transaction_qty']
monthly_sales = df.groupby(df['transaction_date'].dt.to_period('M'))['total_sales'].sum()
percentage_change = monthly_sales.pct_change() * 100
print(percentage_change)
```

```python
plt.figure(figsize=(10, 6))
bar_colors = ['green' if x >= 0 else 'red' for x in percentage_change]
percentage_change.plot(kind='bar', color=bar_colors)
plt.title('Month-on-Month Change in Sales')
plt.xlabel('Month')
plt.ylabel('Percentage Change (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

MONTH ON MONTH TOTAL ORDERS INCREASE/DECREASE

```python
monthly_orders = df.groupby(df['transaction_date'].dt.to_period('M')).size()
print(monthly_orders)
```

```python
percentage_change_orders = monthly_orders.pct_change() * 100
print(percentage_change_orders)
```

```python
plt.figure(figsize=(10, 6))
bar_colors = ['green' if x >= 0 else 'red' for x in percentage_change_orders]
percentage_change_orders.plot(kind='bar', color=bar_colors)
plt.title('Month-on-Month Change in Number of Orders')
plt.xlabel('Month')
plt.ylabel('Percentage Change (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

TOTAL QTY SOLD FOR EACH MONTH

```python
monthly_qty_sold = df.groupby(df['transaction_date'].dt.to_period('M'))['transaction_qty'].sum()
print(monthly_qty_sold)
```

```python
percentage_change_qty_sold = monthly_qty_sold.pct_change() * 100
print(percentage_change_qty_sold)
```

```python
plt.figure(figsize=(10, 6))
bar_colors = ['green' if x >= 0 else 'red' for x in percentage_change_orders]
percentage_change_qty_sold.plot(kind='bar', color=bar_colors)
plt.title('Month-on-Month Change in Number of Orders')
plt.xlabel('Month')
plt.ylabel('Percentage Change (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df['total_sales'] = df['unit_price'] * df['transaction_qty']

df['year_month'] = df['transaction_date'].dt.to_period('M')

monthly_customer_sales = df.groupby(['year_month', 'transaction_id'])['total_sales'].sum().reset_index()

average_order_value = monthly_customer_sales.groupby('year_month')['total_sales'].mean().reset_index()

plt.figure(figsize=(14, 8))
plt.bar(average_order_value['year_month'].astype(str), average_order_value['total_sales'], color='skyblue')

for index, value in enumerate(average_order_value['total_sales']):
    plt.text(index, value, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

plt.xlabel('Month')
plt.ylabel('Average Order Value')
plt.title('Average Order Value per Transaction Month on Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

SALES ANALYSIS ON WEEKDAYS AND WEEKENDS OF MAY

```python
df_may = df[df['transaction_date'].dt.month == 5]
df_may['day_type'] = df_may['transaction_date'].dt.dayofweek.apply(lambda x: 'weekend' if x >= 5 else 'weekday')
df_may['total_sales'] = df_may['unit_price'] * df_may['transaction_qty']
total_sales_by_day_type = df_may.groupby('day_type')['total_sales'].sum()
print(total_sales_by_day_type)
```

```python
plt.figure(figsize=(10, 6))
total_sales_by_day_type.plot(kind='barh', color=['skyblue', 'salmon'])
plt.title('Total Sales in Weekdays and Weekends for May')
plt.xlabel('Total Sales')
plt.ylabel('Day Type')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

TOTAL SALES ANALYSIS BASED ON WEEKDAYS AND WEEKENDS

```python
df['day_type'] = df['transaction_date'].dt.dayofweek.apply(lambda x: 'weekend' if x >= 5 else 'weekday')
df['total_sales'] = df['unit_price'] * df['transaction_qty']
total_sales_by_day_type = df.groupby('day_type')['total_sales'].sum()
print(total_sales_by_day_type)
```

```python
total_sales_by_day_type = df.groupby('day_type')['total_sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.barh(total_sales_by_day_type['day_type'], total_sales_by_day_type['total_sales'], color=['skyblue', 'salmon'])
plt.title('Total Sales in Weekdays and Weekends')
plt.xlabel('Total Sales')
plt.ylabel('Day Type')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
total_sales_by_day_type = df.groupby('day_type')['total_sales'].sum().reset_index()
sns.set(style="whitegrid")
colors = ['#ff9999', '#66b3ff']


plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(
    total_sales_by_day_type['total_sales'],
    labels=total_sales_by_day_type['day_type'],
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},
    textprops={'fontsize': 14},
    pctdistance=0.85
)


for text in texts:
    text.set_fontsize(14)
    text.set_color('black')
    text.set_fontweight('bold')
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')


plt.title('Total Sales Contribution: Weekdays vs Weekends', fontsize=18, fontweight='bold')
plt.tight_layout()
plt.show()
```

DAILY SALES ANALYSIS FOR JANUARY

```python
df_jan = df[df['transaction_date'].dt.month == 1]

df_jan['total_sales'] = df_jan['unit_price'] * df_jan['transaction_qty']
daily_sales_jan = df_jan.groupby(df_jan['transaction_date'].dt.date)['total_sales'].sum().reset_index()


average_sales = daily_sales_jan['total_sales'].mean()


plt.figure(figsize=(14, 7))
plt.plot(daily_sales_jan['transaction_date'], daily_sales_jan['total_sales'], marker='o', label='Daily Sales', color='b')
plt.axhline(average_sales, color='r', linestyle='--', linewidth=2, label=f'Average Sales: {average_sales:.2f}')


for date, sales in zip(daily_sales_jan['transaction_date'], daily_sales_jan['total_sales']):
    if sales > average_sales:
        plt.bar(date, sales, color='green', alpha=0.5)
    else:
        plt.bar(date, sales, color='red', alpha=0.5)


plt.title('Daily Sales for January')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```

SALES ANALYSIS BY PRODUCT CATEGORY

```python
df['total_sales'] = df['unit_price'] * df['transaction_qty']
category_sales = df.groupby('product_category')['total_sales'].sum().reset_index()


category_sales = category_sales.sort_values(by='total_sales', ascending=False)


plt.figure(figsize=(12, 8))
bars = plt.barh(category_sales['product_category'], category_sales['total_sales'], color='skyblue')

for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, f'{width:.2f}', 
             va='center', ha='left', color='black', fontsize=10)

plt.xlabel('Total Sales')
plt.ylabel('Product Category')
plt.title('Total Sales by Product Category')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()  # Invert y-axis to have the highest sales at the top


plt.tight_layout()
plt.show()


```

TOP 10 PRODUCT BY SALES
